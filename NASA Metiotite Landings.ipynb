```python
url = "https://data.nasa.gov/resource/gh4g-9sfh.csv"
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(url)
df
```

```python
df = df[df['year'].notna()]
```

```python
def convert_datetime_to_date(row):
    return row['year'][:4]
```

```python
df['year'] = df.apply(convert_datetime_to_date, axis=1)
df
```

```python
seriesYears = df['year'].value_counts()
```

```python
plt.title('Количество падений метеоритов в год')
plt.ylabel('Год')
plt.xlabel('Количетво метеоритов')
seriesYears[seriesYears > 8].plot.barh()
plt.show()
```
