# Статистика финалов NBA

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel("NBA Finals and MVP.xlsx", sheet_name="Data")
df.head()
```

### Количество чемпионств

```python
seriesChampions = df["NBA Champion"].value_counts()
```

```python
plt.title('Количество чемпионств команд')
seriesChampions.plot.bar()
plt.show()
```

### Соотношение побед команд на западе и востоке

```python
eastChamps = df[df["Eastern Champion"] == df["NBA Champion"]]
westChamps = df[df["Western Champion"] == df["NBA Champion"]]
```

```python
plt.title('Статистика побед команд с запада и востока')
plt.pie([len(eastChamps), len(westChamps)], labels = ("Запад","Восток"))
plt.show()
```

### Самый низкий баскетболист, который получил MVP

```python
df[df["MVP Height (m)"].notna()]["MVP Height (m)"].min()
```
